```python
import pandas as pd
import csv
readframe=pd.read_csv('unrecognized.csv', index_col=False, quoting=csv.QUOTE_ALL)
readframe.Name=readframe.Name.str.strip()
readframe
```

```python
readframe.Name[0][0]==readframe.Name[1][0]
```

```python
ll=[]
skip=0
for x in range(0,227):
    if x<=224 and skip!=1:
        if readframe.Name[x][0]=='A' and readframe.Name[x+1][0]!='A' and readframe.Name[x+2][0]=='A':
            ll.append(readframe.Name[x]+' '+readframe.Name[x+1])
            skip=1
        else:
            ll.append(readframe.Name[x])
    else:
        skip=0
ll
```

```python
ll=[]
skip=0
for x in range(0,1778):
    if x<=1775 and skip!=1:
        if readframe.Name[x][0]==readframe.Name[x+2][0] and readframe.Name[x+1][0]!=readframe.Name[x][0]:
            ll.append(readframe.Name[x]+' '+readframe.Name[x+1])
            skip=1
        else:
            ll.append(readframe.Name[x])
    else:
        skip=0
ll
```

```python
pd.DataFrame({'Name':ll})
```
