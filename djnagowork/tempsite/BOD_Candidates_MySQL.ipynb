```python
import MySQLdb as db
import pandas as pd
from pandas import DataFrame
from pandas.io.sql import frame_query

database = db.connect('localhost','root','yoyo','himanshu')
bodframe     = frame_query("select person_name, name from (select * from company_official limit 88000) c1, company c2 where c1.id=c2.id", database)
bodframe.person_name = bodframe.person_name.str.lower()
bodframe
```

```python
import csv
candframe=pd.read_csv('myneta.2014_againfetched.csv', index_col=False, quoting=csv.QUOTE_ALL)
candframe.Party=candframe.Party.str.lower()
candframe.Candidate=candframe.Candidate.str.lower()
candframe
```

```python
#rename the person_name field of bodframe
bodframe.rename(columns={'person_name':'Candidate'}, inplace=True)
bodframe
```

```python
#let's bare join
#skip this if suing the next cell
res=pd.merge(bodframe, candframe , how='inner', on='Candidate')
mainres=res[['Candidate','name','Constituency','Education','ID','Party','Total Assets']]
#gradres=mainres[mainres.Education=='Graduate']
gradres=mainres
partyres=gradres[gradres.Party=='bjp']
partyres
```

```python
#let's Levensthein join
#but works only for ten cands against a list of bod
import Levenshtein
bodframe['index']=0
candframe['index']=0
candframe2=candframe[candframe.Party=='bjp']
candframe3=candframe2[candframe2.Education=='Graduate']
res=pd.merge((candframe3.head(100)).tail(10) ,bodframe, on='index')


#del candframe['Dummy']
#del bodframe['Dummy']

res['distance'] = res.apply(lambda r: Levenshtein.ratio(r['Candidate_x'], r['Candidate_y']), axis=1)
res[res.distance>0.9][['Candidate_x', 'Candidate_y','name','Constituency','Education','ID','Party','Total Assets']]
```
