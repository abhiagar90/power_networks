```python
#sudo apt-get install python-bs4
#sudo apt-get install python-requests
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq
import re
```

```python

def scrapeCompanyWiki(url):
    r = rq.get(url)
    data = r.text
    
    soup = BeautifulSoup(data)
    
    head = soup.find('h2', attrs={'class':'directorHeading'})
    print head.text
    
    df = pd.DataFrame()
    table = soup.find('table', attrs={'class':"responcive_tablewiki"})
    columns = [] # to get the columns or heading of data
    rows = table.find_all('tr')
    for row in rows:
        headers = row.find_all('th') 
        for h in headers:
            columns.append(h.text)
    
    print columns
    data2 = []
    for row in rows:
        cols = row.find_all('td') 
        if len(cols) > 0:
            dict2 = {columns[0]:cols[0].text , columns[1]:cols[1].text,columns[2]:cols[2].text, columns[3]:cols[3].text,columns[4]:cols[4].text}
            data2.append(dict2)
    
    df  = pd.DataFrame(data2)
    
    return df
```

```python
url = "http://www.companywiki.in/advanceddirectorsearchresult?option=Director&value=ABU+HASEM"
```

```python
alldf = scrapeCompanyWiki(url)
```
