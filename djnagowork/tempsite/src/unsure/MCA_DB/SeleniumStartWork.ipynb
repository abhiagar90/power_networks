```python
'''Assuming the url is good
This method finds company details for the url
Added patch: for replacing " to ' and ; to , to conform to our csv standards
'''
def getCompanySpecificDetails_CW(driver,url):
    log('Now: '+url)
    driver.get(url)
    seen = waitTillIDVisible(driver, 'topSearchInput',140,5)
    if not seen:
        print "Net doesnt work good when "+url
        return None ## adding patch for CIN 
        ##http://www.companywiki.in/company/greater-noida-integrated-warehousing-private-limited%28100%-f.sub.co%29/u63023dl2005ptc140106
    comp = Company()
    comp.name = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[1]/header/h3[1]').text.strip().replace(";",",").replace('"',"'")
    comp.cin = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[1]/header/h3[2]').text.strip().replace(";",",").replace('"',"'")
    comp.startdate = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[6]/table/tbody/tr[2]/td[1]/p').text.strip().replace(";",",").replace('"',"'")
    comp.regnum = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[6]/table/tbody/tr[2]/td[2]/p').text.strip().replace(";",",").replace('"',"'")
    comp.typeofcompany = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[6]/table/tbody/tr[3]/td[1]/p').text.strip().replace(";",",").replace('"',"'")
    comp.listingtype = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[6]/table/tbody/tr[3]/td[2]/p').text.strip().replace(";",",").replace('"',"'")
    comp.busscategory = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[6]/table/tbody/tr[4]/td[1]/p').text.strip().replace(";",",").replace('"',"'")
    ##Patch to avoid the problem with ; in a particular category
    comp.companynature = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[6]/table/tbody/tr[4]/td[2]/p').text.strip().replace(";",",").replace('"',"'")
    comp.subcategory  = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[6]/table/tbody/tr[5]/td[1]/p').text.strip().replace(";",",").replace('"',"'")
    comp.regauth = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[6]/table/tbody/tr[5]/td[2]/p').text.strip().replace(";",",").replace('"',"'")
    comp.authcap = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[7]/table/tbody/tr[2]/td[1]/p').text.strip().replace(";",",").replace('"',"'")
    comp.authcap = comp.authcap.replace('Rs. ','').replace(";",",").replace('"',"'")
    comp.authcap = comp.authcap.replace(',','').replace(";",",").replace('"',"'")
    comp.paidupcap = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[7]/table/tbody/tr[2]/td[2]/p').text.strip().replace(";",",").replace('"',"'")
    comp.paidupcap = comp.paidupcap.replace('Rs. ','').replace(";",",").replace('"',"'")
    comp.paidupcap = comp.paidupcap.replace(',','').replace(";",",").replace('"',"'")
    comp.email = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[11]/table/tbody/tr[4]/td/p').text.strip().replace(";",",").replace('"',"'")
    comp.address  = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[11]/table/tbody/tr[2]/td[1]').text.strip().replace(";",",").replace('"',"'")
    comp.address = comp.address.replace('\n',' ')
    comp.address = comp.address.replace('\r',' ')
    comp.address = comp.address.replace('Registered Address ','')
    comp.address = comp.address.replace('Not Available','').strip().replace(";",",").replace('"',"'") ##always came with the tail, got rid of it
    #/html/body/div[1]/div[4]/article/div[4]/h3[2]
    comp.status = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[4]/h3[2]').text.strip().replace(";",",").replace('"',"'")
    comp.status = comp.status[comp.status.find(":")+1:].strip()
    ##Added patch after http://www.companywiki.in/company/vippy-industries-limited/l15142mp1973plc001225
    #print comp.status 
    dirlist = []
    dirs = driver.find_element_by_xpath('/html/body/div[1]/div[4]/article/div[8]/table')
    trs = dirs.find_elements_by_tag_name("tr")
    if len(trs[2].find_elements_by_tag_name("td"))==4:
        for tr in trs[2:]: ##skipping first two
            tds = tr.find_elements_by_tag_name("td")
            persondir = Director()
            persondir.din = tds[0].text.strip().replace(";",",").replace('"',"'")
            #print persondir.din
            persondir.name = tds[1].text.strip().replace(";",",").replace('"',"'")
            persondir.startdate =tds[2].text.strip().replace(";",",").replace('"',"'")
            persondir.desig = tds[3].text.strip().replace(";",",").replace('"',"'")
            dirlist.append(persondir)
    comp.dirlist =dirlist
    return comp
#print getCompanySpecificDetails_CW(driver,'http://www.companywiki.in/company/greater-noida-integrated-warehousing-private-limited%28100%-f.sub.co%29/u63023dl2005ptc140106')
```
