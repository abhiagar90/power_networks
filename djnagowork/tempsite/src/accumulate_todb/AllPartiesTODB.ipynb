```python
def adddots(partyname):
    toreturn=''
    bracketoff=True
    for x in partyname:
        if ((x>='A' and x<='Z') or (x>='a' and x<='z')) and bracketoff:
            toreturn+=(x+'.')
        elif x=='(':
            toreturn+=x
            bracketoff=False
        elif x==')':
            toreturn+=x
            bracketoff=True
        else:
            toreturn+=x        
    return toreturn
somedots=adddots('cpi')
if(somedots[-1]=='.'):somedots=somedots[:-1]
somedots
```
