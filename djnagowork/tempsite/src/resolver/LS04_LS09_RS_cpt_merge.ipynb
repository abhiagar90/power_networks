```python
import preprocess
import join_df
import pandas as pd
import MySQLdb as db
import numpy as np
from pandas.io.sql import frame_query
```

```python
ls2004 = pd.read_csv('./Myneta_scrape - ls2004_winners.csv')
ls2009 = pd.read_csv('./Myneta_scrape - ls2009_winners.csv')
ls2004
```

RS dataframe must be formatted so that all NaNs are empty strings

```python
rs_dependencies = pd.read_csv('./RS_members_relatives.csv')
rs_dependencies = rs_dependencies.replace(np.nan,'', regex=True)
rs_dependencies['spouse'][0]
```

Time for some preprocessing - We bring all the relevant attributes of join in a uniform format.  
This includes - 
1. ls04 candidate names  
2. ls09 candidate names  
3. rs_member dependency names

```python
ls2004['Candidate'] = ls2004.apply(lambda r : preprocess.name_preprocess(r['Candidate']),axis = 1)
ls2004
```

```python
ls2009['Candidate'] = ls2009.apply(lambda r : preprocess.name_preprocess(r['Candidate']),axis = 1)
ls2009
```

```python
rs_dependencies['father'] = rs_dependencies.apply(lambda r : preprocess.name_preprocess(r['father']), axis =1)
rs_dependencies['mother'] = rs_dependencies.apply(lambda r : preprocess.name_preprocess(r['mother']), axis =1)
rs_dependencies['spouse'] = rs_dependencies.apply(lambda r: preprocess.name_preprocess(r['spouse']), axis =1)
rs_dependencies
```

Taking in the dir table dataframe with same formatiing.

```python
database = db.connect('localhost','root','yoyo','c1')
dirframe = frame_query("select * from dir_table_amartya", database)
dirframe
```

```python
dirframe['person_name'] = dirframe.apply(lambda r : preprocess.name_preprocess(r['person_name']) , axis =1)
dirframe
```

```python
from time import time
start = time()
df_merged = join_df.custom_merge(ls2004,dirframe,'Candidate','person_name',0.88)
end = time()
print "join performed in {} seconds".format(end - start)
```

```python
df_merged.to_csv('LS04_cpt.csv')
df_merged
```
