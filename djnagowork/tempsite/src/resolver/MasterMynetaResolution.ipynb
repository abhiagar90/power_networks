```python
import pandas as pd
import csv
partyframe=pd.read_csv('partyids.csv', index_col=False, quoting=csv.QUOTE_ALL)
partyframe.Party=partyframe.Party.str.lower()
partyframe.Party=partyframe.Party.str.strip()
partyframe
```

```python
lswinners=pd.read_csv('ls2014winners.csv', index_col=False, quoting=csv.QUOTE_ALL)
lswinners.Party=lswinners.Party.str.lower()
lswinners.Party=lswinners.Party.str.strip()
lswinners.State=lswinners.State.str.lower()
lswinners.State=lswinners.State.str.strip()
lswinners.Constituency=lswinners.Constituency.str.lower()
lswinners.Constituency=lswinners.Constituency.str.strip()
lswinners
```

```python
lswinners_1 = pd.merge(lswinners, partyframe, on='Party')
lswinners_1.drop('Party', axis=1, inplace=True)
lswinners_1
```

```python
stateframe=pd.read_csv('stateids.csv', index_col=False, quoting=csv.QUOTE_ALL)
stateframe.State=stateframe.State.str.lower()
stateframe.State=stateframe.State.str.lower()
stateframe
```

```python
lswinners_2 = pd.merge(lswinners_1, stateframe, on='State')
#lswinners_1.drop('Party', axis=1, inplace=True)
lswinners_2
```

```python
constframe=pd.read_csv('consids.csv', index_col=False, quoting=csv.QUOTE_ALL)
constframe.Constituency=constframe.Constituency.str.lower()
constframe.Constituency=constframe.Constituency.str.strip()
constframe.State=constframe.State.str.lower()
constframe.State=constframe.State.str.strip()
constframe
```

```python
lswinners_3 = pd.merge(lswinners_2, constframe, on=['State','Constituency'])
#lswinners_1.drop('Party', axis=1, inplace=True)
lswinners_3
```

```python
lswinners_3.drop('State', axis=1, inplace=True)
lswinners_3.drop('Constituency', axis=1, inplace=True)
```

```python
lswinners_3.to_csv('lswinners2014_resolved.csv', index=False)
```
