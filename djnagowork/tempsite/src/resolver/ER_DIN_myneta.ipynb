Get the similarity scores as a dictionary

```python
def get_cosine_sim(A,B):
    
    A = A.strip().lower()
    B = B.strip().lower()
    AB_list = [A,B]
    v = TfidfVectorizer()
    
    tfidf = v.fit_transform(AB_list)
    
    return (tfidf*tfidf.T).A

    
    
def get_similarity_scores(A,B):
    
    A = A.strip().lower()
    B = B.strip().lower()
    
    leven = Levenshtein.ratio(A,B)
    jaro = jf.jaro_distance(A,B)
    fuzzy = fuzz.ratio(A,B)
    
    return {'lev':leven, 'jar' : jaro, 'fuz':fuzzy}


def is_match(A,B,th):
    
    sim_scores = get_similarity_scores(A,B)
    
    if sim_scores['jar'] > 0.6 or sim_scores['fuz'] >80 :
        return True
    else:
        return False
```

####The address matching function goes below.  
First tried to follow a heuristic for similarity (jaro and fuzzy ) between total addr string and statename

The second function divides address into tokens and tries to match using exact state name or a jaro distance of it.This function is used. 

```python
#heuristic 1
def extract_states(addr):
    
    state_list = ['Andhra Pradesh', 'Arunachal Pradesh','Assam', 'Bihar', 'Chhattisgarh', 'Goa' ,'Gujarat' ,'Haryana' ,'Himachal Pradesh', 'Jammu and Kashmir' ,'Jharkhand', 'Karnataka' ,'Kerala', 'Madhya Pradesh' ,'Maharashtra', 'Manipur' ,'Meghalaya' ,'Mizoram' ,'Nagaland','Orissa' ,'Punjab' ,'Rajasthan', 'Sikkim','Tamil Nadu', 'Tripura', 'Uttarakhand', 'Uttar Pradesh', 'West Bengal','Telangana'] # update to all states in india
    state_list = [s.lower() for s in state_list]
    ut_list = ['Andaman and Nicobar Islands' ,'Chandigarh' ,'Dadar Nagar Haveli' ,'Daman and Diu', 'Delhi' ,'Lakshadweep' ,'Pondicherry']
    ut_list = [u.lower() for u in ut_list]
     
    max_jaro = -1
    max_fuzz = -1
    
    stateObtained = False
    state  = ''
    for s in state_list:
        
        jaro = jf.jaro_distance(s,addr)
        
        fuzzval = fuzz.token_set_ratio(s,addr)
        
        if jaro > max_jaro or fuzzval > max_fuzz :
            state = s
            max_jaro = jaro
            max_fuzz = fuzzval
    
    print 'state obtained '+state
    
    return state


#heuristic 2
def extract_states_ut_exact(addr):
    
    state_list = ['Andhra Pradesh', 'Arunachal Pradesh','Assam', 'Bihar', 'Chhattisgarh', 'Goa' ,'Gujarat' ,'Haryana' ,'Himachal Pradesh', 'Jammu and Kashmir' ,'Jharkhand', 'Karnataka' ,'Kerala', 'Madhya Pradesh' ,'Maharashtra', 'Manipur' ,'Meghalaya' ,'Mizoram' ,'Nagaland','Orissa' ,'Punjab' ,'Rajasthan', 'Sikkim','Tamil Nadu', 'Tripura', 'Uttarakhand', 'Uttar Pradesh', 'West Bengal','Telangana'] # update to all states in india
    state_list = [s.lower() for s in state_list]
    ut_list = ['Andaman and Nicobar Islands' ,'Chandigarh' ,'Dadar Nagar Haveli' ,'Daman and Diu', 'Delhi' ,'Lakshadweep' ,'Pondicherry']
    ut_list = [u.lower() for u in ut_list]
    
    s = ''
    for s in state_list:
        
        addr_list = addr.split(',')
        
        addr_list = [a.strip().lower() for a in addr_list]
        for a in addr_list:
            
            if (s == a) or (jf.jaro_distance(s,a)) > 0.85:
                
                return s
        
    for u in ut_list:
        
        addr_list = addr.split(',')
        
        addr_list = [a.strip().lower() for a in addr_list]
        for a in addr_list:
            
            if (u == a) or (jf.jaro_distance(u,a)) > 0.85:
                
                return u
        
    print 'No state found for -'+addr
    return ''
```

Custom join ( no optimization) . Nested loop join using custom condition! This is to alleviate problem 3 above.

```python

def custom_merge_brute(df1 , df2):
    
    col1 = list(df1.columns)
    col2 = list(df2.columns)
    
    df_new = pd.DataFrame(columns = col1+col2)
    
    index_new = 0
    
    for index,row in  df1.iterrows():
        for index_inner,row_inner in df2.iterrows():
            
            key_df1 = df1['Candidate'][index]
            key_df2 = df2['name'][index_inner]
            key2_df1 = df1['State'][index]
            key2_df2 = df2['state'][index_inner]
            
            if key_df1 == key_df2 or jf.jaro_distance(key_df1,key_df2) >= 0.88:
                #print 'yay1'
                #print key_df1
                #print key_df2
                #print key2_df1
                #print key2_df2
                if key2_df1 == key2_df2 or jf.jaro_distance(key2_df1,key2_df2) >= 0.9:
                    
                    #print 'yay'
                    #print row
                    #print row_inner
                
                    #get new values of df_new
                    dfnew_cols = list(df_new.columns)
                    
                    tmp_dict = {'Candidate':[df1['Candidate'][index]],'ID':[df1['ID'][index]],'State':[df1['State'][index]],'profession':[df1['profession'][index]],'din':[df2['din'][index_inner]],'name':[df2['name'][index_inner]],'address':[df2['address'][index_inner]],'state':[df2['state'][index_inner]]}
                    tmp_df = pd.DataFrame(tmp_dict)
                    df_new = pd.concat([df_new,tmp_df],ignore_index = True)
                    
    
    
    return df_new         

df_merge5 = custom_merge_brute(ls2014_req_1,dirframe_req)
```
