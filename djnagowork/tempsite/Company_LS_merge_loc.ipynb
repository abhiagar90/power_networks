###Merge now---

```python
def custom_merge_brute(df1 , df2):
    
    col1 = list(df1.columns)
    col2 = list(df2.columns)
    
    df_new = pd.DataFrame(columns = ['Party','Candidate','Candidate ID','State','position','dir_name','dir_id','dir_state'])
    
    index_new = 0
    
    for index,row in  df1.iterrows():
        for index_inner,row_inner in df2.iterrows():
            
            key_df1 = df1['Candidate'][index]
            key_df2 = df2['person_name'][index_inner]
            key2_df1 = df1['State'][index]
            key2_df2 = df2['state'][index_inner]
            
            if key_df1 == key_df2 or jf.jaro_distance(key_df1,key_df2) >= 0.8:
                #print 'yay1'
                #print key_df1
                #print key_df2
                #print key2_df1
                #print key2_df2
                if key2_df1 == key2_df2 or jf.jaro_distance(key2_df1,key2_df2) >= 0.9:
                    
                    #print 'yay'
                    #print row
                    #print row_inner
                
                    #get new values of df_new
                    dfnew_cols = list(df_new.columns)
                    
                    tmp_dict = {'Candidate':[df1['Candidate'][index]],'Party':[df1['Party'][index]],'Candidate ID':[df1['Candidate ID'][index]],'State':[df1['State'][index]],'dir_name':[df2['person_name'][index_inner]],'dir_state':[df2['state'][index_inner]],'dir_id':[df2['id'][index_inner]],'position':[df2['designation'][index_inner]]}
                    tmp_df = pd.DataFrame(tmp_dict)
                    df_new = pd.concat([df_new,tmp_df],ignore_index = True)
                    
    
    
    return df_new 
```
